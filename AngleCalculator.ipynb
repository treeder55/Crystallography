```python
# Calculates the angles between high symmetry lattice planes for any arbitrary crystal structure
class angles:
    def __init__(self, varss):
        self.real = varss[0];
        self.ang = varss[1];
        a=self.real[0]; b=self.real[1]; c=self.real[2]; Alpha=self.ang[0]*2*np.pi/360; Beta=self.ang[1]*2*np.pi/360; Gamma=self.ang[2]*2*np.pi/360; #angles converted to radians
        self.vol = a*b*c*np.sqrt(1-(np.cos(Alpha))**2-(np.cos(Beta))**2-(np.cos(Gamma))**2+2*np.cos(Alpha)*np.cos(Beta)*np.cos(Gamma));
        self.Gstar = (2*np.pi)**2/(self.vol**2)*np.array([[b**2*c**2*np.sin(Alpha)**2, a*b*c**2*(np.cos(Alpha)*np.cos(Beta) - np.cos(Gamma)), a*b**2*c*(np.cos(Alpha)*np.cos(Gamma) - np.cos(Beta))], [a*b*c**2*(np.cos(Alpha)*np.cos(Beta) - np.cos(Gamma)), a**2*c**2*np.sin(Beta)**2, a**2*b*c*(np.cos(Beta)*np.cos(Gamma) - np.cos(Alpha))], [a*b**2*c*(np.cos(Alpha)*np.cos(Gamma) - np.cos(Beta)), a**2*b*c*(np.cos(Beta)*np.cos(Gamma) - np.cos(Alpha)), a**2*b**2*np.sin(Gamma)**2]]);
        # Gstar is the reciprocal space metric tensor
        self.highsymm = np.array(varss[2]);
        self.highsymmstr = np.zeros(len(self.highsymm)).astype(str)
        for i in range(len(self.highsymm)):
            self.highsymmstr[i] =str(self.highsymm[i])
        self.table = np.round(np.arccos(np.divide(np.inner(self.highsymm,np.dot(self.highsymm,self.Gstar)),np.sqrt(np.tensordot(np.diag(np.inner(self.highsymm,np.dot(self.highsymm,self.Gstar))),np.diag(np.inner(self.highsymm,np.dot(self.highsymm,self.Gstar))),axes=0))))*360/(2*np.pi),decimals=5);
    def ttable(self):
        self.astropytable = Table(self.table,names = self.highsymm);
        self.astropytable.add_column(Column(self.highsymmstr,name = 'Angle (\N{DEGREE SIGN})'),index=0);
        #print(astropytable)
       # for i in range(len(self.highsym)):
       #     self.table[i] = np.divide(np.multiply(self.highsymm[i],self.highsymm) 
  #  def crossproduct(v1,v2,self):
  #      self.cp = [v1/]
```
